# Color palettes using Seaborn

This notebook uses several palettes defined in https://seaborn.pydata.org/tutorial/color_palettes.html.

Included is a function to print the palettes for use with TikZ/PGFPlots.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
num_colors = 6
# palette = "BuGn_r"
palette = "cubehelix" # Good for printing
# palette = "Set2"
# palette = "Paired"
# palette = sns.hls_palette(6, l=.47, s=.8)
# palette = sns.cubehelix_palette(num_colors, start=.5, rot=-.75)
# palette = sns.cubehelix_palette(num_colors, start=2, rot=0, dark=0, light=.85, reverse=True)
# palette = sns.cubehelix_palette(num_colors)
# palette = sns.diverging_palette(220, 20, n=7)
# palette = sns.diverging_palette(10, 220, sep=10, n=num_colors)
sns.palplot(sns.color_palette(palette, num_colors))
plt.show()
```

## Plots
You can plot the colors of the palette here

```python
sns.set_palette(palette)
for i in range(num_colors):
    plt.plot([0, 1], [0, i+1])

sns.set_style("whitegrid")
plt.show()
```

## TikZ/PGFPlots
You can also print the color to define them for use with TikZ and PGF plots.

```python
def print_colors(colors, indices):
    counter = 0
    for i in indices:
        counter += 1
        print('\definecolor{mycolor' + str(counter) + '}{rgb}{' + str(colors[i][0]) + ',' + str(colors[i][1]) + ',' + str(colors[i][2]) + '}')

colors = sns.color_palette(palette, num_colors)
indices = [0,1,2] # Set the indices of the colors you want to use from your palette. Index starts counting at 0

print_colors(colors, indices)
```
